```python
#read the data from csv files 
import pandas as pd
local_data = pd.read_csv('local_data.csv')
global_data = pd.read_csv('global_data.csv')
```

```python
local_data['ma_10'] = local_data['avg_temp'].rolling(10).mean()
global_data['ma_10'] = global_data['avg_temp'].rolling(10).mean()
```

```python
# diffrence between local and global average temperature
df_diff = pd.DataFrame(global_data['year'])
df_diff['diff'] = local_data['avg_temp'] - global_data['avg_temp']
# moving average for the diffrence
df_diff['ma'] = df_diff['diff'].rolling(10).mean()
```
